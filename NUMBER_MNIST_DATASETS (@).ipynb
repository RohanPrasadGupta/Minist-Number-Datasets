```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
(x_train,y_train),(x_test,y_test) = keras.datasets.mnist.load_data()
```

```python
print('x_train shape: ',x_train.shape)
print('x_test shape: ',x_test.shape)
print('y_train shape: ',y_train.shape)
print('y_test shape: ',y_test.shape)
```

```python
#Arranging the value between 0 and 1
x_train = x_train/255
x_test = x_test/255
```

```python
print('x_train shape: ',x_train.shape)
print('x_test shape: ',x_test.shape)
print('y_test shape: ',y_test.shape)
```

```python
x_test[0]
```

```python
x_train_flattend = x_train.reshape(len(x_train),28*28)
```

```python
x_train_flattend.shape
```

```python
x_test_flattend = x_test.reshape(len(x_test),28*28)
print(x_test_flattend.shape)
print(y_test.shape)
```

```python
model = keras.Sequential([
    keras.layers.Dense(10,input_shape=(784,), activation='sigmoid' )
])
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']

)
```

```python
model.fit(x_test_flattend,y_test, epochs=10)
```

```python
model.evaluate(x_test_flattend,y_test)
```

```python
weig, bias = model.get_weights()
```

```python
print(weig)
print("Weight Shape: ",weig.shape)
```

```python
cm= tf.math.confusion_matrix(labels=y_test, predictions=y_predict_table)
cm
```
